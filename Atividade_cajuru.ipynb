# **Qual é o hospital/upa mais próximo e o caminho mais curto?**

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

input_data = pd.read_csv('data_cajuru.csv', index_col=0)

G = nx.Graph(input_data.values)

# Calcular o caminho mais curto usando o algoritmo de Dijkstra
source_node = 0
target_node = 45

# Verifica se os nós de origem e destino estão presentes no grafo
if source_node not in G.nodes():
    print("Nó de origem não encontrado no grafo.")
elif target_node not in G.nodes():
    print("Nó de destino não encontrado no grafo.")
else:
    shortest_path = nx.dijkstra_path(G, source=source_node, target=target_node)

    shortest_path_graph = G.subgraph(shortest_path)

    # Plotar o grafo do caminho mais curto
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=5, node_color='skyblue', font_size=10, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=shortest_path_graph.edges(), width=2, edge_color='red')
    plt.title("Caminho mais curto representado no grafo  de casa até o hospital")
    plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

input_data = pd.read_csv('data_cajuru.csv', index_col=0)
G = nx.Graph(input_data.values)

node_names = dict(enumerate(input_data.columns))

# Calcular o caminho mais curto usando o algoritmo de Dijkstra
source_node = 0
target_node = 45

if source_node not in G.nodes():
    print("Nó de origem não encontrado no grafo.")
elif target_node not in G.nodes():
    print("Nó de destino não encontrado no grafo.")
else:
    shortest_path = nx.dijkstra_path(G, source=source_node, target=target_node)
    shortest_path_length = nx.dijkstra_path_length(G, source=source_node, target=target_node)

    # Obtenha as informações de distância entre cada par de nós no caminho mais curto
    path_distances = [(shortest_path[i], shortest_path[i + 1], G[shortest_path[i]][shortest_path[i + 1]]['weight']) for i in range(len(shortest_path) - 1)]
    total_distance = sum(distance for _, _, distance in path_distances)

    # Criar um novo grafo contendo apenas os nós e as arestas do caminho mais curto
    shortest_path_graph = G.subgraph(shortest_path)

    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, labels=node_names, with_labels=True, node_size=5, node_color='skyblue', font_size=5, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=shortest_path_graph.edges(), width=2, edge_color='red')
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{G[u][v]['weight']} metros" for u, v, _ in path_distances}, font_size=6)
    plt.title(f"Caminho mais curto representado no grafo  de casa até o hospital (Distância total: {total_distance} metros)")
    plt.show()
```
